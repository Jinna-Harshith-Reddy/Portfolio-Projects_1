```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tracks = pd.read_csv('C:/Users/Welcome/Desktop/spotify_python_project/tracks.csv') #reads the csv file from the path
df_tracks.head() #shows first 5 rows in the dataset
```

```python
#null values check

pd.isnull(df_tracks).sum()
```

```python
#information regarding each colmn

df_tracks.info()
```

```python
# Finding 10 least popular artists

sorted_df = df_tracks.sort_values('popularity', ascending = True).head(10)

sorted_df
```

```python
# for descriptive statistics

df_tracks.describe().transpose()
```

```python
# Finding top 10 songs with popularity > 90

most_popular = df_tracks.query('popularity > 90', inplace = False).sort_values('popularity', ascending = False)
most_popular[:10]  #slices the data to show the first 10 rows
```

```python
# creating a new index 'release_date'

df_tracks.set_index('release_date', inplace = True)  #changes the index
df_tracks.index = pd.to_datetime(df_tracks.index)   #changes the format to datetime
df_tracks.head()
```

```python
# Filters the data and checks which artist is present in the 18th row of the dataset

df_tracks[["artists"]].iloc[18]
```

```python
# Converting durations from milliseconds to seconds

df_tracks["duration"] = df_tracks["duration_ms"].apply(lambda x: round(x/1000))  #uses lambda function to divide x by 1000
df_tracks.drop("duration_ms", inplace = True, axis = 1)  #drops the duratin_ms column in the table
```

```python
# Check for duration

df_tracks.duration.head()
```

```python
# Visualisation for correlation between different columns

corr_df = df_tracks.drop(["key", "mode", "explicit"], axis = 1).corr(method = "pearson") #drops those columns and applies pearson correlation method
plt.figure(figsize=(24,6))  #plots the figure
heatmap = sns.heatmap(corr_df, annot = True, fmt = ".1g", vmin = -1, vmax = 1, center = 0, cmap = "inferno", linewidth = 1, linecolor = "Black") #creates a heatmap, cmap means colormap for the vis. rest are self explanatory
heatmap.set_title("Correlation HeatMap Between Variables")  #gives title to the heatmap
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation = 90) #rotation 90deg, rotates the horizontal labels by 90deg
```

```python
# Regression plots

sample_df = df_tracks.sample(int(0.004*len(df_tracks)))  #creates a sample data frame
```

```python
print(len(sample_df))  #0.4% of total dataset. These many rows we have in our sample data frame
```

```python
#There was a positive correlation between loudness and enery in heatmap, so let's plot regression for that

plt.figure(figsize = (10, 6))  #sets the figure size
sns.regplot(data = sample_df, y = "loudness", x = "energy", color = "c").set(title = "Loudness vs Energy Correlation")  #plots the regression
```

```python
# Regression plot for popularity and acousticness

plt.figure(figsize = (10, 6))  #sets the figure size
sns.regplot(data = sample_df, y = "popularity", x = "acousticness", color = "b").set(title = "Popularity vs Acousticness Correlation")  #plots the regression
```

```python
# Creating new column: Year from release_date

df_tracks['dates'] = df_tracks.index.get_level_values('release_date')
df_tracks.dates = pd.to_datetime(df_tracks.dates)
years = df_tracks.dates.dt.year
```

```python
# Creating a visualisation for total number of songs in each year since 1982

sns.displot(years, discrete = True, aspect = 2, height = 5, kind = "hist").set(title = "Number of songs per year")  #hist=>histogram plot
```

```python
# Year vs Duration bar plot

total_dr = df_tracks.duration
fig_dims = (18, 7)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x = years, y = total_dr, ax = ax, errwidth = False).set(title = "Year vs Duration")
plt.xticks(rotation = 90)
```

```python
# Year vs Duration line plot

total_dr = df_tracks.duration
sns.set_style(style = "whitegrid")
fig_dims = (10, 5)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.lineplot(x = years, y = total_dr, ax = ax).set(title = "Year vs Duration")
plt.xticks(rotation = 60)
```
